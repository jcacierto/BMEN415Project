```python
#import libraries
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn import *
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
```

Exploratory Data Analysis

```python
df = pd.read_csv('breast cancer.csv')
x = df.copy()
x = x.drop(['diagnosis', 'id'], axis=1)
y = df["diagnosis"]

y = y.replace(to_replace="M", value=0)
y = y.replace(to_replace="B", value=1)

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 555)
#print(df)
print('Shape: ', df.shape)
print('--------------------')
print(df.dtypes)
print('--------------------')
print(df.head)

```

**Linear Discriminant Analysis**

```python
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

#perform LDA
lda = LDA(n_components = 1)
x_train = lda.fit_transform(x_train, y_train)
x_test = lda.transform(x_test)

#evaluate performance using Random Forest classifier
classifier = RandomForestClassifier(max_depth=2, random_state=555)
classifier.fit(x_train, y_train)
y_pred_train = classifier.predict(x_train)
y_pred_test = classifier.predict(x_test)
```

```python
#Performance Evaluation
cm = confusion_matrix(y_test, y_pred_test)
sns.heatmap(cm, annot = True)

LDA_accuracy_train = accuracy_score(y_train, y_pred_train)
LDA_accuracy_test = accuracy_score(y_test, y_pred_test)

print('Train dataset accuracy for LDA: ', LDA_accuracy_train)
print('Test dataset accuracy for LDA: ', LDA_accuracy_test)
```

**Decision Tree**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 555)

dt = DecisionTreeClassifier()
dt = dt.fit(x_train, y_train)

y_pred_train = dt.predict(x_train)
y_pred_test = dt.predict(x_test)

DT_accuracy_train = accuracy_score(y_train, y_pred_train)
DT_accuracy_test = accuracy_score(y_test, y_pred_test)

print('Train dataset accuracy for Decision Tree: ', DT_accuracy_train)
print('Test dataset accuracy for Decision Tree: ', DT_accuracy_test)\

cm = confusion_matrix(y_test, y_pred_test)
sns.heatmap(cm, annot = True)
```

**Neural Networks - Multi-Layer Perceptron**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 555)

#Making the MLP model
mlp = MLPClassifier(hidden_layer_sizes=(8,8,8), activation='relu', solver='adam', max_iter=500)
mlp.fit(x_train,y_train)

y_pred_train = mlp.predict(x_train)
y_pred_test = mlp.predict(x_test)

MLP_accuracy_train = accuracy_score(y_train, y_pred_train)
MLP_accuracy_test = accuracy_score(y_test, y_pred_test)

print('Train dataset accuracy: ', MLP_accuracy_train)
print('Test dataset accuracy: ', MLP_accuracy_test)

cm = confusion_matrix(y_test, y_pred_test)
sns.heatmap(cm, annot = True)
```
